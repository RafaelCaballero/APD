```python
modules = ["bs4", "spacy", "pandas","seaborn","scipy"]

import sys
import os.path
from subprocess import check_call
import importlib
import os

def instala(modules):
  print("Instalando módulos")
  for m in modules:
      # para el import quitamos [...] y ==...
      p = m.find("[")
      mi = m if p==-1 else m[:p]
      p = mi.find("==")
      mi = mi if p==-1 else mi[:p]

      torch_loader = importlib.util.find_spec(mi)
      if torch_loader is not None:
          print(m," encontrado")
      else:
          print(m," No encontrado, instalando...",end="")  
          try:        
            r = check_call([sys.executable, "-m", "pip", "install", "--user", m])
            print("¡hecho!")
          except:
            print("¡Problema al instalar ",m,"! ¿seguro que el módulo existe?",sep="")
              
  print("¡Terminado!")

instala(modules) 
```

**Ejercicio 1** Cargar el texto de la página:

https://www.gutenberg.org/cache/epub/8870/pg8870.html  mediante requests (para obtener) y BeautifulSoup (para analizar y extraer el texto) dejándolo en una variable de nombre `texto`. Asegurarse de que requests no da error, mostrando un mensaje si el `status_code` es distinto de 200

```python
from bs4 import BeautifulSoup
import requests
pagina = "https://www.gutenberg.org/cache/epub/8870/pg8870.html"

page = requests.get(pagina)
if page.status_code==200:
    soup = BeautifulSoup(page.text, 'html.parser')  # le pasamos el texto en HTML para que lo analice
    texto = soup.text
else:
    print("Error ", page.status_code)
```

**Ejercicio 2** Aplicar el modelo `nlp`al `texto` dejando el resultado en una variable `doc`.
A partir de `doc` extraer una lista de todos los lexemas en mayúsculas de los tokens que 

1) Sean valores alfauméricos
2) No correspondan a palabras vacías
3) Sean sustantivos

El resultado será una lista de palabras en mayúscula (lexemas) que guardaremos en una variable `palabras`

Ayuda: un token es un sustantivo si `token.pos_ == "NOUN"`. La palabra *pos* viene de *part of speech* e indica el tipo de palabra que es. Ver la [ayuda de spaCy](https://spacy.io/usage/linguistic-features) para un ejemplo

**Ejercicio 5** Escribir una función:

Nombre `palabrasEnFrase`

Parámetros:

    palabras: una lista de strings representando palabras
    frase: un string representando una frase
   
Salida: una lista de 0s y 1s, con la misma longitud que `palabras` tal que en la posición i tendrá un 1 si la palabra iésima aparece en la frase y 0 en otro caso


Ayuda: `hola in 'me dijo hola'` devuelve `True`, y `hola in 'se lo llevó la ola'` devuelve `False`

```python
def palabrasEnFrase(palabras,frase):
    # solución
    return [ 0 if p not in frase else 1 for p in palabras]

# salida esperada: [1, 0, 1, 0, 1, 0]
palabrasEnFrase(["PRINCESA","RAMA", "RANA", "PRINCIPE","POZO", "GOZO"],
                "HABÍA UNA VEZ UNA LINDA RANA QUE AL CAERSE A UN POZO SE CONVIRTIÓ EN UNA HORRIBLE PRINCESA")
```

**Ejercicio 8** Las matrices de 0 y 1s se pueden considerar por un lado tipo ratio y por otro lado nominales. Vamos a considerar ahora que todas las columnas son nominales y en lugar de calcular las correlaciones numéricas vamos a utilizar las asociaciones entre variables nominales. 

Es decir, ahora para cada par de columnas tendremos que calcular su tabla de contingencias y el test $\mathcal{X}^2$. Lo único que vamos a ser muy exigentes: solo descartamos la hipóstesis nula y consideramos que hay una posible relación cuando `p< 10e-8`. Mostrar las parejas distintas que verifican que hay relación entre ellas con estas condiciones 

Nota: evitar que salgan parejas de columnas repetidas o una columna consigo misma

Comentario general: saldrán muy parecidas al ejercicio anterior pero con alguna nueva. LA desventaja es que no vemos si es asociación positiva o negativa, por lo que se suele preferir el método anterior en el caso de variables "dummy". En el caso de variables nominales es muy habitual la secuencia:
1) Buscar asociaciones con  $\mathcal{X}^2$ entre variables nominales
2) Convertir a variables dummy con one-hot-encoding
3) Repetir las correlaciones, ahora ya como variables numéricas con `corr()`que nos mostrará posibles correlaciones con otras variables ratio y entre sí, pero ahora entre valores individuales de la varaible nominal



```python
from scipy.stats import chi2_contingency
for c1 in df2.columns: 
    for c2 in df2.columns:
         if c1<c2:
            contingencias=pd.crosstab(index=df2[c1],columns=df2[c2])
            ChiSqResult = chi2_contingency(contingencias)
            p = ChiSqResult[1]
            if p<10e-8:
                print(c1,c2, ". p: ",round(p,5))
```

**Ejercicio 9** 

Nombre: `bigramas`

Parámetros: `l` , una lista de strings o palabras

Devuelve: El resultado de concatenar cada palabra con la siguiente si son distintas con un "-" entremedias

Por ejemplo, si `l=["hola","qué", "tal", "te", "te", "va"]`, tendremos como salida `['hola-qué', 'qué-tal', 'tal-te', 'te-va']`


```python
def bigramas(l):
    return [l[i]+"-"+l[i+1] for i in range(len(l)-1) if l[i]!=l[i+1]]

bigramas(["hola","qué", "tal", "te",  "te", "va"])
```

**Ejercicio 10** (largo) Ahora queremos relacionar bigramas de palabras unos con otros (es decir, 4 elementos, 2 con 2). Para ello vamos a repetir buena parte del código de la práctica pero adaptada 

1) aplicamos bigramas a palabras, lo que da una variable bi

2) Contamos los N elementos más frecuentes en bi, a los que podemos llamar `bifrecuentes`

3) Escribimos una función `bigramasEnFrase`, que dada una lista de bigramas y una frase devuelve una lista de 1s  y 0s indicando qué bigramas están en una frase. Nota: Un bigrama está en una frase si están las dos palabras que lo componen

4) Aplicamos la función con `bifrecuentes`y cada frase en `frases` lo que nos dará una lista de listas `bipertenencia`

5) Con `bipertenencia` creamos un dataframe `df_bi`, del que quitamos las filas que tengan todo 0s dando lugar a `df2_bi`

6) Calculamos las correlaciones en  `df2_bi` y mostramos las parejas de bigramas que tengan una correlación mayor que 0.10 en valor absoluto (rebajamos la exigencia de 0.25 a 0.10). Solo debemos mostrar parejas de bigramas que no tengan ninguna palabra en común, es decir deben representar 4 palabras distintas.

```python
# solución
bi = bigramas (palabras)

N=25
c = Counter(bi)
bifrecuentes = [p for p,_ in c.most_common(N)]

def bigramasEnFrase(bigramas,frase):
    # solución
    return [ 0 if (p.split("-")[0] not in frase) or (p.split("-")[1] not in frase) else 1 for p in bigramas]

bipertenencia = [bigramasEnFrase(bifrecuentes,f) for f in frases]


df_bi = pd.DataFrame(bipertenencia,columns=bifrecuentes)

df2_bi = df_bi[df_bi!=0].dropna(how="all").fillna(0).reset_index(drop=True)

correlaciones = df2_bi.corr()
for c1 in df2_bi.columns:
    for c2 in df2_bi.columns:
        if c1<c2 and set(c1.split("-")).intersection(c2.split("-"))==set():
            v = correlaciones.loc[c1,c2]
            if abs(v)>0.10:
                print(c1,c2,v)
```
