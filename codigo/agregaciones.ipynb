## Groupby




Consideramos el siguiente dataframe:

```python
import pandas as pd
datos = [
    { "nombre": "Bertoldo", "mes": "Marzo", "distKm": 6,  "tiempoMin": 42,  "comp": False},
    { "nombre": "Herminia", "mes": "Marzo", "distKm": 10, "tiempoMin": 60,  "comp": True},
    { "nombre": "Bertoldo", "mes": "Marzo", "distKm": 2,  "tiempoMin": 12,  "comp": False},
    { "nombre": "Herminia", "mes": "Marzo", "distKm": 10, "tiempoMin": 61,  "comp": True},
    { "nombre": "Bertoldo", "mes": "Abril", "distKm": 5,  "tiempoMin": 33,  "comp": False},
    { "nombre": "Herminia", "mes": "Abril", "distKm": 42, "tiempoMin": 285, "comp": True},
    { "nombre": "Aniceto",  "mes": "Abril", "distKm": 5,  "tiempoMin": 33,  "comp": False},
]

df = pd.DataFrame(datos)
df
```

Nosotras sabemos ya calcular la distancia media:

```python
df["distKm"].mean()
```

Pero ¿y si queremos la distancia media **por nombre**? Una opción sería dividir el dataframe en varios, uno por nombre distinto y calcular allí las medias. Por fortuna no tenemos que hacer esto; Pandas ya se encarga de hacerlo por nosotros 

```python
media_por_nombre = df.groupby("nombre")["distKm"].mean()
media_por_nombre
```

Lo que hace esta instrucción es:

1) Crea los grupos con `df.groupby("nombre")`

```python
df.groupby("nombre")
```

```python
for nombre, g in df.groupby("nombre"):
    print("\n=== Grupo:", nombre, "===")
    print(g)
```

Ahora seleccionamos la columna que queremos y aplicamos la función de agregación que deseemos

```python
df.groupby("nombre")["distKm"].mean()
```

Con dos columnas: 

```python
media_por_nombre = df.groupby("nombre")[["distKm", "tiempoMin"]].mean()
media_por_nombre
```

¿Y qué ocurre si queremos aplicar un función diferente según cada columna? 

En este caso usaremos `agg`

## Multinivel

Ejercicio: seleccionar el `tiempoMin` máximo por nombre y mes

```python
 df.groupby(["nombre","mes"])[ "tiempoMin"].max()
```

```python
res = df.groupby(["nombre","mes"])[ ["tiempoMin"]].max()
res
```

El resultado es un índice multinivel o "multiindex". La selección se sigue pudiendo hacer con loc:

```python
res.loc[("Bertoldo", "Marzo"),:]    
```

```python
res.loc[(slice(None), "Marzo"),:]   
```

El `slice(None)` sustituye al : en índices multinivel

## pivot_table

[pivot_table](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html) Añade una función de agregación a pivot.

```python
stocks = pd.read_csv('https://gist.githubusercontent.com/alexdebrie/b3f40efc3dd7664df5a20f5eee85e854/raw/ee3e6feccba2464cbbc2e185fb17961c53d2a7f5/stocks.csv')
stocks
```

Vamos a convertir la columna `symbol`en 3, una para cada uno de sus valores

```python
df_symbol = stocks.pivot(index='date', columns='symbol', values='volume')
df_symbol
```

¿Y qué sucede si el índice se repite? Pues que pivot_table hace utiliza una función de agregación, por defecto la suma

```python
stocks.pivot_table(index='symbol', values='volume')
```

Ahora añadimos una función de agregación de forma explícita

```python
import numpy as np
stocks.pivot_table(index='symbol', values='volume', aggfunc="sum")
```

```python
import numpy as np
stocks.pivot_table(index='symbol', values='volume', aggfunc="mean")
```
