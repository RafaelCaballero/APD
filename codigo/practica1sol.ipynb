a) Construir un diccionario con nombre `datos`y que tenga 3 campos: `s`, `n`, `a`, cada campo tomando respectivamente `sexo`, `nombres`, `àño`

**Pregunta 9** 

La variable `acabaena` tendrá True en todas las posiciones que correspondan a nombres acabados en "a". Ejecutar le siguiente código para inicializarla

```python
acabaena = np.char.endswith(nombres, "a")

nombres[acabaena][:5] # 5 primeros nombres terminados en a
```


Ahora queremos obtener la proporción de  chicas (sexo "girl") cuyo nombre acaba en `a` y la proporción de chicos (sexo "boy") cuyo nombre acaba en 'a'. Dejaremos los resultados en dos variables, `pchicasa` y `pchicosa`




```python
# solución
chico = sexo=='boy'
chica = sexo=='girl'


pchicosa = (chico & acabaena).sum() / chico.sum() # en estadística es P(acabar en "a" | chico)  (la | se lee "condicionado a")
pchicasa = (chica & acabaena).sum() / chica.sum() # P(acabar en "a" | chica)

# No cambiar
print(f"Proporción de chicos cuyo nombre acaba en 'a': {pchicosa:.4f}, y prop. de chicas cuyo nombre acaba en 'a': {pchicasa:.4f} ")
```

**Pregunta 10** Asumiendo que en 'sexo' solo hay dos valores, "boy" o "girl", ¿Qué frase de la siguientes te parece más correcta al hablar de intentar deducir el sexo de la persona según si su nombre acaba en 'a' o no? Selecciona la opción que te parezca más acertada a partir de los resultados anteriores, y escribe una frase que explique tu respuesta

    a) El método parece útil para decidir con alta probabilidad  si es chico o chica a partir el nombre
    b) El método parece útil para asegurar con alta probabilidad que es chica si acaba en 'a'
    c) El método solo parece útil para asegurar con alta probabilidad  que es chico si no acaba en 'a'


Solución: Editar esta casilla con doble click 

b) porque la probabilidad de que sea chico acabando en 'a' es mínima 

Si queremos podemos verlo aplicando el teorema de Bayes:

Si llamo A a "acabar en a" tenemos que la probabilidad de que sea chico si sabemos que acaba en A es:

$$
P(chico | A) = \frac{P(chico)P(A|chico)}{P(A)}
$$

$$
P(chica | A) = \frac{P(chica)P(A|chica)}{P(A)}
$$

```python
# probabilidad de que siendo chico acabe en a
print(f"P(chico|A): {chico.mean() * pchicosa / acabaena.mean() : .4f}")
print(f"P(chica|A): {chica.mean() * pchicasa / acabaena.mean() : .4f}")
```
