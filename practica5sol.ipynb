```python
modules = ["geopy","folium","pprint"]


import sys
import os.path
from subprocess import check_call
import importlib
import os

def instala(modules):
    print("Instalando módulos")
    for m in modules:
        # para el import quitamos [...] y ==...
        p = m.find("[")
        mi = m if p==-1 else m[:p]
        p = mi.find("==")
        mi = mi if p==-1 else mi[:p]
        torch_loader = importlib.util.find_spec(mi)
        if torch_loader is not None:
            print(m," encontrado")
        else:
            print(m," No encontrado, instalando...",end="")  
            try:        
                r = check_call([sys.executable, "-m", "pip", "install", "--user", m])
                print("¡hecho!")
            except:
                print("¡Problema al instalar ",m,"! ¿seguro que el módulo existe?",sep="")

    print("¡Terminado!")

instala(modules)  
```

En ocasiones las funciones no devuelven nada, no tienen `return`. Cuando esto sucede es porque la función tiene "efectos laterales": por ejemplo crear una carpeta o, como en este caso, mostrar por pantalla el contenido de una carpeta

```python

# lista los ficheros y subcarpetas del path
def lista(path):
    print(path)
    espacio = " "*5
    for p in path.iterdir():
        if p.is_dir():
            print(espacio,p.name,": carpeta",sep="")
        else:
            if p.is_file():
                print(espacio,p.name,": fichero",sep="")
    print("="*40)
                
#### llamada
lista(path)
```

*Nombre*: `descarga(path,url)`

*Salida*: no devuelve nada, solo descarga usando requests el fichero indicado

Pista: en la solución de la práctica 4, ejercicio 2, podemos ver un ejemplo de descarga

```python
import requests

# solución
def descarga(path,url):
    r = requests.get(url, allow_redirects=True) # el fichero queda en la variable r
    if r.status_code == 200:
        with open(path, 'wb') as f:
            f.write(r.content) # ahora lo grabamos localmente
    else:
        print("Error ",r.status_code)

# no tocar a partir de aquí, es para probar descargando dos ficheros, clientes.csv y gestores.csv que vamos a usar a continuación
url_clientes = "https://raw.githubusercontent.com/RafaelCaballero/tdm/master/datos/clientes2.csv"  
path_raw = Path(path,"raw")
path_clientes = Path(path_raw,"clientes.csv")
descarga(path_clientes,url_clientes)
path_gestores = Path(path_raw,"gestores.csv")
url_gestores = "https://raw.githubusercontent.com/RafaelCaballero/tdm/master/datos/clientegestor.csv"
descarga(path_gestores,url_gestores)
lista(path_raw)
```

*Nombre*: `reparte(path,df)`
- `path`: el path de una carpeta donde se alojarán los nuevos ficheros creados a partir de df, uno por gestor
- `df`: el dataframe a repartir en varios, debe incluir una columna `gestor`que se usara para repartir

*Salida*: no devuelve nada, solo descarga usando requests el fichero indicado

Idea: para cada valor diferente de la columna gestor (usar un `for`) se usara yun filtro para las filas del gestor. Se aplicará el filtro para obtener un dataframe nuevo con los datos del gestor, que a continuación se grabará en la carpeta indicada

```python
# solución
def reparte(path,df):
    for g in df.gestor.unique():
        print(g)
        path_g = Path(path,g+".csv")
        filtro_g = df.gestor==g
        df_g  = df[filtro_g]
        df_g.to_csv(path_g,index=False)
        
## llamada         
path_repartidos = Path(path,"repartidos")
reparte(path_repartidos,df_gc)
# para comprobar
lista(path_repartidos)
```

### map

El método anterior para crear una columna nueva puede llegar a ser muy complicado; por ello se suele usar `map`. Vamos a verlo para hacer algo análogo con la columna T2

La idea es la siguiente: primero se escribe una función en Python que devuelva el valor nuevo de la columna T2 para un valor concreto de T (es la idea de siempre, pensar en números y luego extender a columnas/filas)

```python
def posneg(t):
    if t<0:
        res=-1
    elif t==0:
        res=0
    else:
        res=+1
    return res

# para probar
print("-0.85 --> ",posneg(-0.85))
print("+0.15 --> ",posneg(+0.15))
print("+0 --> ",posneg(+0))

        
```


Una vez que hemos visto que funciona ya podemos crear la nueva columna T2 que será el resultado de aplicar la función `posneg` a cada valor de la columna T, uno a uno. La función `map` de `Series`nos permite hacer eso

```python
df_usa["T2"] = df_usa.trump.map(posneg)
df_usa
```

### apply 
`map` es muy útil, pero a veces resulta insuficiente. Supongamos que ahora en df_usa queremos añadir una columna "voto" que indique si el usuario es "Pro-trump" (trump>clinton), "pro-clinton" (clinton<trump) o "neutro". El problema es que ahora no podemos usar map porque se aplica a una columna y nuestra nueva columna depende de 2, trump y clinton. 

Por fortuna `apply` nos permite aplicar una función a una fila entera.

```python
df_usa
```

```python
def pro(row): 
    #print(row)
    if row["trump"]>row["clinton"]:
        res="Pro-trump"
    elif row["trump"]<row["clinton"]:
        res="Pro-clinton"
    else:
        res="neutro"
    return(res)
        
pro(df_usa.iloc[0])        
```

Aplicamos la función

```python
df_usa["voto"] = df_usa.apply(pro,axis=1) # obsérvese que ahora no se aplica a una columna sino al dataframe entero
df_usa
```
